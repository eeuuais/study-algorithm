## 1. 넘파이(Numpy) : 배열(ndarray, n dimension array)

- 대표적인 두 자료형(리스트, 넘파이)의 비교
- 리스트(list): Linked List 구조(불연속), 동작 속도가 느리다, comma로 요소 구별, 서로 다른 자료형 사용가능
- (넘파이 활용한) 배열(ndarray): 연속된 메모리 구조, 동작 속도가 빠르다, comma를 사용하지 않고 공백으로 요소를 구별, 오직 한 가지 타입
만 갖을 수 있다

** 어디서 주로 사용할까? **

```python
import numpy as np
import matplotlib.pylab as plt
from sklearn.datasets import load_digits

digits = load_digits()
samples = [0, 10, 20, 30, 1, 11, 21, 31]

print(type(digits.images))
print(digits.images[100])

d = []
for i in range(8):
    d.append(digits.images[samples[i]])
    
# print(d)

plt.figure(figsize=(8, 2))
for i in range(8):
    plt.subplot(1, 8, i + 1)
    plt.imshow(d[i], interpolation='nearest', cmap=plt.cm.bone_r)
    # plt.imshow(digits.images[900], interpolation='nearest', cmap=plt.cm.bone_r)
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    plt.title("image {}".format(i + 1))
plt.suptitle("number Image")
plt.tight_layout()
plt.show()
```

```python
from scipy import misc

img_rgb = misc.face()
print(img_rgb.shape)

print(img_rgb)

plt.subplot(221)
plt.imshow(img_rgb, cmap=plt.cm.gray) # 컬러 이미지 출력
plt.axis("off")
plt.title("RGB image")

plt.subplot(222)
plt.imshow(img_rgb[:, :, 0], cmap=plt.cm.gray) # red 채널 출력
plt.axis("off")
plt.title("Red channel")

plt.subplot(223)
plt.imshow(img_rgb[:, :, 1], cmap=plt.cm.gray) # green 채널 출력
plt.axis("off")
plt.title("Green channel")

plt.subplot(224)
plt.imshow(img_rgb[:, :, 2], cmap=plt.cm.gray) # blue 채널 출력
plt.axis("off")
plt.title("Blue channel")

plt.show()
```

```python
m1 = np.arange(12)
m3 = m1.reshape(2,2,3)
print(m3,m3.shape) # (2,2,3)

m4 = m1.reshape(3,2,1,2)
print(m4,m4.shape) # (3,2,1,2)

m = np.arange(18).reshape(3,6)
print(m,m.shape)
```

```python
# -1을 사용한 reshape
m1 = np.arange(6)
print(m1,m1.shape) # (6,)

m2 = m1.reshape(-1,2) # 행의 요소수를 자동으로 계산, 열은 2열로 반드시 맞추어라
print(m2,m2.shape) # (3,2)

m2 = m1.reshape(-1,3) # 열의 요소수를 자동으로 계산, 행은 반드시 3행으로 맞추어라
print(m2,m2.shape) # (2,3)
```

```python
m2 = m1.reshape(-1,2) # 행의 요소수를 자동으로 계산, 열은 1개로 만들어라
print(m2,m2.shape) # (6, 1)
```

```python
m3 = m1.reshape(-1,2,3) # 면의 요소수를 자동으로 계산, 2*3행렬을 우선 만들어라
print(m3,m3.shape) # (1,2,3)

m3 = m1.reshape(-1,3,2) # 면의 요소수를 자동으로 계산, 3*2행렬을 우선 만들어라
print(m3,m3.shape) # (1,3,2)

m3 = m1.reshape(2,3,-1) # 면의 요소수를 자동으로 계산, 2개 큰 행렬을 먼저 만들고 3행짜리를 만들어라 그 뒤의 열은 신경안쓴다.
print(m3,m3.shape) # (2,3,1)

m4 = np.arange(18)
print(m4,m4.shape) # (18,)
m4 = m4.reshape(2,3,-1) # 면의 요소수를 자동으로 계산, 2개 큰 행렬을 먼저 만들고 3행짜리를 만들어라 그 뒤의 열은 신경안쓴다.
print(m4,m4.shape) # (2,3,3)

m5 = np.arange(18)
print(m5,m5.shape) # (18,)
m5 = m5.reshape(3,-1,2) # 면의 요소수를 자동으로 계산, 3개 큰 행렬을 먼저 만들고 가운데는 임의로, 마지막은 2로.
print(m5,m5.shape) # (3,3,2)
```

### 4. 배열의 인덱싱과 슬라이싱

```python
# 1차원 배열의 인덱싱과 슬라이싱 : list 와 동일한ㅂ 방법
d = np.arange(6)
print(d,d.shape)
# indexing
print(d[0],d[1],d[2],d[-1],d[-2])
# slicing
print(d[2:3],d[:],d[:-1])
print(d[::2],d[::-1])
d[0] = 10 # list 처럼 요소 변경이 가능
print(d)
```

```python
# 2차원 배열의 인덱싱과 슬라이싱
# data[행,열]
d = np.arange(12).reshape(3,4)
print(d,d.shape) # (3, 4)

# indexing
print(d[0][0]) # 0
print(d[0,0]) # 0
print(d[2,1]) # 9

# slicing
print(d[0]) # [0 1 2 3]
print(d[0,:]) # [0 1 2 3]
print(d[:,:-1]) # 마지막 열을 제외 ,(3,3)
print(d[::2,::2]) # (2,2)ㅑ
d[0] = 20 # 0번 행의 모든 요소를 20으로 변경
print(d)

# d5[::2,::3] = 30 # 슬라이싱 된 요소 전체를 한번에 변경할 수 있다
print(d[::2,::3])
print(d)

d[:,:] = -1 # 전체 요소를 한번에 변경할 수 있다
print(d)
```

```python
# 3차원 배열의 인덱싱과 슬라이싱

# data[면,행,열]
d = np.arange(12).reshape(2,2,3)
print(d,d.shape)

# indexing
print(d[0][0][0]) # old style
print(d[0,0,0]) # 0
print(d[0,1,2]) # 5

# slicing
print('-'*30)
print(d[:,:-1,:-1]) # (2, 1, 2)
print('-'*30)
print(d[:,1:,1:]) # (2, 1, 2) , 3-D
print('-'*30)
print(d[:,:,-1]) # (2, 2), matrix
print('-'*30)
print(d[:,-1,-1]) # (2,), vecotr
print(d[-1,-1,-1]) # 11, scalar
```

```python
# 인덱싱 결과는 scalar
# 슬라이싱 결과는 ndarray
a = np.arange(12)
print(a,a.shape)

print(a[0]) # 0 , scalar
print(a[0:1]) # [0] , 배열(ndarray)
```

- 필터를 활용하여 필터에 True 조건만 색인
- 꺾쇠 괄호로 한 번 더 묶어만 주면 됩니다. -n - (꺾쇠 괄호 안에 조건문 삽입)

```python
a[a>2]
```

```python
a[a!=3]
```

```python
print(a[(a > 2) & (a < 6)])
print(a[a % 2 == 0])
print(a[a < a.mean()]) # 배열 a에서 평균보다 작은 값만 추출
```

```python
print(a[True])
print(a[False])
```

## 10. View, Copy

### 1) View
- View : 원본 객체와 메모리를 공유하는 사본, 원본과 id가 다름
- Copy : 원본 객체와 다른 별도의 메모리를 사용하는 사본, 원본과 id가 다름

```python
import numpy as np
# 1 View
a = np.array([2,8,9,7,3,1,4,5,6])
# v = a
v = a.view() # v = a 와 동일한 결과
a,v
```

```python
v[0] = 100
a[1] = 200
print(v,a)
```

```python
print(id(v),id(a))
print(np.shares_memory(v,a)) # 메모리를 공유하므로 True가 반환
```

### 2) Copy

```python
# 2 Copy
c = a.copy()
c[0] = 500
a[1] = 600
print(a,c)

print(id(a),id(c))
print(np.shares_memory(c,a)) # 메모리를 공유하지 않으므로 False가 반환
```

```python
# 리스트를 슬라이싱한 객체는 원본과 메모리를 공유하지 않음
a = [0,1,2,3,4,5]
b = a[:3]
print(a,b)
b[0] = 100
print(a,b)
print(np.shares_memory(a,b)) # False
```

```python
# 배열을 슬라이싱한 객체는 원본과 메모리를 공유
a = np.array([0,1,2,3,4,5])
b = a[:3]
print(a,b)
b[0] = 100
print(a,b)
print(np.shares_memory(a,b)) # True
print(id(a),id(b))
```

- 수직(v,행)으로 합치기 : vstack(), row_stack(), concatenate(..., axis=0)
- 수평(h,열)으로 합치기 : hstack(), column_stack(), concatenate(..., axis=1)

### 1) 수직으로 합치기

```python
# 수직으로 합치기
a = np.arange(9).reshape(3,3)
b = a * 2
print(a)
print(b)
```

```python
r = np.vstack((a,b)) # 수직으로 열은 그대로 행을 붙여서 합친다.
print(r,r.shape)
```

```python
x = np.concatenate((a,b),axis=0) # axis=0, 수직, 행을 기준으로
print(x,x.shape)
```

```python
r = np.hstack((a,b)) # 수평
print(r,r.shape)

c = np.column_stack((a,b)) # 열
print(c,c.shape)

x = np.concatenate((a,b),axis=1) # axis=1, 수평, 열
print(x,x.shape)
```

## 3) 브로드 캐스팅 : 차원의 크기가 서로다른 배열에서 산술 연산이 가능

- 브로드캐스팅(Broadcasting)은 모양이 다른 배열들 간의 연산이 어떤 조건을 만족했을 때 가능해지도록 배열을 자동적으로 변환

### 브로드캐스팅이 일어날 수 있는 조건
- 1) 차원의 크기가 1일때
- 2) 두 배열 간의 연산에서 최소한 하나의 배열의 차원이 1이라면(0번 축이든 1번 축이든; 1행이든 1열이든) 가능
- 3) 차원의 짝이 맞을 때 가능
- 4) 차원에 대해 축의 길이가 동일하면 가능

- 규칙 1: 두 배열의 차원 수가 다르면 1차원 데이터를 배열(행벡터, 열벡터)형태로 즉, shape의 앞쪽(왼쪽)/행(열)을 1(차원형태가 나오게)이
되도록 만든다.
- 규칙 2 : 차원이 같게되면, 즉 행이나 열이 같은 경우, 행벡터(혹은 열벡터)로 만들어진 배열(shape 이 1인 배열)이 다른 배열의 shape과 일치
하도록 늘인다
- 규칙 3 : 임의의 차원에서 크기가 일치하지 않으면 오류가 발생

- 브로드 캐스팅 규칙 : 차원과 shape 이 자동 조정
- 규칙 1 : 두 배열의 차원 수가 다르면 더 작은 수의 차원을 가진 배열의 shape의 앞쪽(왼쪽)을 1로 채운다
예 1) A는 2*2
B는 1*2

```python
# 규칙 1 적용 : B의 차원이 정해져 있지 않음. 행벡터로 바꾼다.
# B.shape ==> (1,3) , [[0 1 2]]
# 계속 B를 그 아래로 같은 값을 A행열의 행과 맞추면서 늘여놓는다.
# 규칙 2 : 차원이 같은 두 배열의 shape 이 1인 배열을 다른 배열의 shape과 일치하도록 늘인다
# 규칙 2 적용 : B.shape ==> (3,3) , [[0 1 2]
# [0 1 2]
# [0 1 2]]
# 그러나 A는 3*2행렬이다.
# 규칙 3 : 임의의 차원(행이나 열에서 같지 않고) 에서 크기가 일치하지 않고 (A의 행렬의 shape가 행열중 어떤 것도 1이 아니라면 오류가 발생
# 규칙 3 에 해당하므로 오류 발생 : 열의 크기가 다르고 shape의 1의 값이 없다
print(A + B) #ValueError: operands could not be broadcast together with shapes (3,2) (3,)
```

### 배열 연산 샘플

```python
import numpy as np
arr1 = np.array([10, 11, 12, 13, 14, 15])
arr2 = np.array([20, 21, 22, 23, 24, 25])
                 
newarr = np.add(arr1, arr2)
print(newarr)
```

```python
arr1 = np.array([10, 20, 30, 40, 50, 60])
arr2 = np.array([20, 21, 22, 23, 24, 25])
newarr = np.subtract(arr1, arr2)
print(newarr)
```

```python
arr1 = np.array([10, 20, 30, 40, 50, 60])
arr2 = np.array([20, 21, 22, 23, 24, 25])
```

```python
newarr = np.multiply(arr1, arr2)
print(newarr)
```

```python
arr1 = np.array([10, 20, 30, 40, 50, 60])
arr2 = np.array([3, 5, 10, 8, 2, 33])
newarr = np.divide(arr1, arr2)
print(newarr)
```

```python
arr1 = np.array([10, 20, 30, 40, 50, 60])
arr2 = np.array([3, 5, 6, 8, 2, 33])
newarr = np.power(arr1, arr2)
print(newarr)
```

```python
arr1 = np.array([10, 20, 30, 40, 50, 60])
arr2 = np.array([3, 7, 9, 8, 2, 33])
newarr = np.mod(arr1, arr2)
print(newarr)
```

```python
arr1 = np.array([10, 20, 30, 40, 50, 60])
arr2 = np.array([3, 7, 9, 8, 2, 33])
newarr = np.remainder(arr1, arr2)
print(newarr)
```

```python
arr1 = np.array([10, 20, 30, 40, 50, 60])
arr2 = np.array([3, 7, 9, 8, 2, 33])
newarr = np.divmod(arr1, arr2)
print(newarr)
```

```python
arr = np.ceil([-3.1666, 3.6667])
print(arr)
arr = np.floor([-3.1666, 3.8667])
print(arr)
```

```python
arr = np.arange(1, 10)
print(np.log2(arr))
arr = np.arange(1, 10)
print(np.log10(arr))
arr = np.arange(1, 10)
print(np.log(arr))
```

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([1, 2, 3])
newarr = np.add(arr1, arr2)
print(newarr)
arr1 = np.array([1, 2, 3])
arr2 = np.array([1, 2, 3])
newarr = np.sum([arr1, arr2])
print(newarr)
```

```python
arr = np.array([1, 2, 3])
newarr = np.cumsum(arr)
print(newarr)
```

```python
arr = np.array([1, 2, 3, 4])
x = np.prod(arr)
print(x)
```

```python
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([5, 6, 7, 8])
x = np.prod([arr1, arr2])
print(x)
```

```python
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([5, 6, 7, 8])
newarr = np.prod([arr1, arr2], axis=1)
print(newarr)
```

```python
arr = np.array([5, 6, 7, 8])
newarr = np.cumprod(arr)
print(newarr)
```

```python
#E.g. for [1, 2, 3, 4], the discrete difference would be [2-1, 3-2, 4-3] = [1, 1, 1]
arr = np.array([10, 15, 25, 5])

newarr = np.diff(arr)
print(newarr)
```

```python
# Finding LCM (Lowest Common Multiple)
num1 = 4
num2 = 6
x = np.lcm(num1, num2)
print(x)
```

```python
arr = np.array([3, 6, 9])
x = np.lcm.reduce(arr)
print(x)
```

```python
#Finding GCD (Greatest Common Denominator)
num1 = 6
num2 = 9

x = np.gcd(num1, num2)
print(x)
```
