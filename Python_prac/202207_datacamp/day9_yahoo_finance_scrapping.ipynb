```python
# 현재 날짜를 기준으로 과거데이터는 lag으로 표시함

time_lags = 1

# 스트링 값을 받아서 컬럼명 생성. t-1 기 데이터를 가져온다
df_lag['Close_Lag(t-%s)' % str(time_lags)] = df['Close'].shift(time_lags) 


# 변화율
# df_lag['Close_Lag(t-%s)_change' % str(time_lags)] = df_lag['Close_Lag(t-%s)' % str(time_lags)].pct_change()*100.0
df_lag['Close_Change'] = df_lag['Close'].pct_change()*100.0


# 방향성
# numpy의 sign함수 이용
df_lag['Close_Direction'] = np.sign(df_lag['Close_Change'])
df_lag
```

```python
# Volume에 대해서도 새로운 컬럼 생성하여 추가

time_lags = 1
df_lag['Volume_Lag(t-%s)' % str(time_lags)] = df['Close'].shift(time_lags) 
df_lag['Volume_Change'] = df_lag['Volume'].pct_change()*100.0
df_lag['Volume_Direction'] = np.sign(df_lag['Volume_Change'])
df_lag
```

- 이것을 함수로 만들어서 사용해보자 

```python
def make_dataset(df, time_lags):
    df_lag = pd.DataFrame(index=df.index)
    df_lag['Close'] = df['Close']
    df_lag['Volume'] = df['Volume']
    
    df_lag['Close_Lag(t-%s)' % str(time_lags)] = df['Close'].shift(time_lags) 
    df_lag['Close_Change'] = df_lag['Close'].pct_change()*100.0
    df_lag['Close_Direction'] = np.sign(df_lag['Close_Change'])
    
    df_lag['Volume_Lag(t-%s)' % str(time_lags)] = df['Close'].shift(time_lags) 
    df_lag['Volume_Change'] = df_lag['Volume'].pct_change()*100.0
    df_lag['Volume_Direction'] = np.sign(df_lag['Volume_Change'])
    return df_lag
```

```python
def get_date_by_percent(start_date, end_date, percent):
    days =  (end_date - start_date).days
    target_days = np.trunc(days * percent)
    target_date = start_date + datetime.timedelta(days=target_days)
    target_date.strftime('%Y-%m-%d')
    return target_date
```

```python
split_ratio = 0.75
percent = split_ratio
split_date = get_date_by_percent(start_date, end_date, split_ratio)
split_date
```

```python
# 위를 함수로 정의한다

def split_dataset(df, input_columns_array, output_column, split_ratio):
    split_date = get_date_by_percent(start_date, end_date, split_ratio)
    #split_date = get_date_by_percent(df.index[0], df.index[df.shape[0]-1, split_ratio])
    input_data = df[input_column_array]
    output_data = df[output_column]

    # 트레이닝, 테스트 셋 생성
    X_train = input_data[input_data.index < split_date]
    X_test = input_data[input_data.index >= split_date]
    Y_train = output_data[output_data.index < split_date]
    Y_test = output_data[output_data.index >= split_date]
    
    return X_train, X_test, Y_train, Y_test
```
