```python
# gtts 패키지로 음성 파일을 만들어보자

from gtts import gTTS

text = "안녕하세요. 석박사 논문 작성을 위한 데이터 분석 과정입니다."
tts = gTTS(text=text, lang='ko')
tts.save("tts_out.mp3")
```

```python
from playsound import playsound

playsound("tts_out.mp3")
```

```python
# 한글 텍스트 파일 내용을 음성으로 출력해보자

from gtts import gTTS
from playsound import playsound

file_path = 'tts_input.txt'
with open(file_path, 'rt', encoding='UTF8') as f : 
    read_file = f.read()
    f.close()

tts = gTTS(text=read_file, lang='ko')
tts.save("tts_file.mp3")
playsound("tts_file.mp3")
```
