```python
# QR코드 여러 개 생성

file_path = "qrlist.txt"
with open(file_path, 'rt', encoding='UTF8') as f : 
    read_lines = f.readlines()
    f.close()
    
    for line in read_lines:
        line = line.strip()   # 공백을 지워준다
        print(line)
        qr_data = line
    qr_img = qrcode.make(qr_data)

    save_path = qr_data + '.png'
    qr_img.save(save_path)
```

```python
# 두 개의 그래프 그리기


import matplotlib.pyplot as plt
import numpy as np

xpoints = np.arange(1, 10, 0.1)
ypoints1 = xpoints*0.2
ypoints2 = np.sin(xpoints)
ypoints3 = np.cos(xpoints)
ypoints4 = np.tanh(xpoints)

plt.plot(xpoints, ypoints1)
plt.plot(xpoints, ypoints2) # 추가할 때마다 그래프 색이 바뀜
plt.plot(xpoints, ypoints3)
plt.plot(xpoints, ypoints4)
plt.show()
```

```python
# 범례 출력


import matplotlib.pyplot as plt
import numpy as np

xpoints = np.arange(1, 10, 0.1)
ypoints1 = xpoints*0.2
ypoints2 = np.sin(xpoints)

plt.plot(xpoints, ypoints1, label = 'first')
plt.plot(xpoints, ypoints2, label = 'second')
plt.legend(loc='upper right')
plt.show()
```

```python
# ColorMap

import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='BuPu')

plt.colorbar()

plt.show()
```

→ 0부터 100까지 가중치에 대해 데이터가 어떻게 흩어져있는지 확인 가능.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes, alpha=0.5) # 투명도 조절

plt.show()
```
