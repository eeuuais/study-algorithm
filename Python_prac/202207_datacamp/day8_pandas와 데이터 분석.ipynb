Anonymous function lambda

펑션명까지 지정하지 않고 1회성으로 함수를 만들어서 쓰고 없앤다.
(아래) 람다에 아규먼트 x를 집어 넣으면 얘는 값이 x+2가 되는거에요

```python
test = lambda x: x+2
test(2)
# range(5) #0,1,2,3,4
```

```python
# Survived를 살았으면 (1) good 죽었으면(0) bad로 대체
# (참일때는 bad 부분이, 거짓일 때는 good 부분이 실행된다)

test = lambda x : "bad" if (x == 0) else "good"

titanic_url['Survived'] = titanic_url['Survived'].map(test)

titanic_url['Survived']
titanic_url[:]
```
