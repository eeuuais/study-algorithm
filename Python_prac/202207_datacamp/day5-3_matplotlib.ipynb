### Matplotlib Plotting

- plot() 함수는 다이어그램에서 점(마커)을 그리는 데 사용
- 포인트를 지정하기 위한 매개변수 사용
- 매개변수 1은 x축의 점을 포함하는 배열
- 매개변수 2는 y축의 점을 포함하는 배열

```python
# pyplot

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0,2,4,6])
ypoints = np.array([0,8,120,250])

# 구간 확대 및 축소
plt.xlim(-2, 9)   # 확대
plt.ylim(0, 300)

# x, y 라벨 및 타이틀
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('matplotlib sample')

# 마커 속성
plt.plot(xpoints, ypoints, marker='o', ms=20, mec='r', mfc='g') # marker edge color

#배경 그리드
plt.grid(color = 'red', linestyle = '--', linewidth = 0.2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.arange(1,10,0.1) # 1에서 10까지 0.1포인트 단위로
ypoints1 = xpoints*0.2
ypoints2 = np.sin(xpoints)
ypoints3 = np.cos(xpoints)
ypoints4 = np.tan(xpoints)

plt.plot(xpoints, ypoints1)
plt.plot(xpoints, ypoints2)
plt.plot(xpoints, ypoints3)
plt.plot(xpoints, ypoints4)
plt.show()
```

```python
# 범례 표시 (legend 출력)

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.arange(1,10,0.1) # 1에서 10까지 0.1포인트 단위로
ypoints1 = xpoints*0.2
ypoints2 = np.sin(xpoints)

plt.plot(xpoints, ypoints1, label='first')
plt.plot(xpoints, ypoints2, label='second')
plt.legend(loc='upper left')
plt.show()
```

### annotation 기능
- 어노테이션은 그래프에  화살표를 그린후에 문자열을 출력하는 기능
- plt.annotate(‘hello’,xy1=(2,10),xy2=(5,20),arrowprops={'color':'green'})

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.arange(1,10,0.1)
ypoints1 = xpoints*0.2
ypoints2 = np.sin(xpoints)

plt.plot(xpoints, ypoints1, label='first')
plt.plot(xpoints, ypoints2, label='second')

# xy = 화살표의 촉 좌표 / xytest = 화살표 뒷부분
plt.annotate('annotate',xy=(2,1),xytext=(4,1.5),arrowprops={'color':'green'})

plt.show()
```

### subplot 기능

- 여러 개의 그래프를 표시하고자 할 때 subplot을 사용
- 그래프가 그려질 위치를 격자로 지정함 : plt.subplot(nrow,ncol,pos)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 10)
y1 = x*5
y2 = x*1
y3 = x*0.3
y4 = x*0.2

plt.subplot(1,2,1) # 로우가 1개, 컬럼이 2개인데 첫번째 칸에 출력하겠다는 것
plt.plot(x,y1)
plt.subplot(1,2,2) # 로우가 1개, 컬럼이 2개인데 두번째 칸에 출력하겠다는 것
plt.plot(x,y2)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 10)
y1 = x*5
y2 = x*1
y3 = x*0.3
y4 = x*0.2

# 그래프 크기 지정
plt.figure(figsize=(10,5))


plt.subplot(2,2,1)
plt.plot(x,y1)
plt.subplot(2,2,2)
plt.plot(x,y2)
plt.subplot(2,2,3)
plt.plot(x,y3)
plt.subplot(2,2,4)
plt.plot(x,y4)

plt.show()
```
